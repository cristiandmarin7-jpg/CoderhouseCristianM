```python
#@title Acciones de limpieza del dataset

#Eliminar columnas poco informativas o con muchos nulos
df.drop(columns=['in_shazam_charts'], inplace= True) # This column is already missing.

# Imputar valores faltantes en la columna 'key' con el valor más frecuente (moda)
if 'key' in df.columns:
    df['key'].fillna(df['key'].mode()[0], inplace=True)
else:
    print("Column 'key' not found in DataFrame.")

print("\nValores faltantes después de la limpieza:")
print(df.isna().sum().sort_values(ascending=False))

df.head()
```

```python
#@title Aplicar algun metodo de Outliers Detection

original_size = df.shape[0]
print(f"Tamaño original del dataset: {original_size}")
```

```python
#@title Con IQR (Interquartile Range)
```

```python
# Convertir 'streams' a numérico (remover comas)
df["streams"] = pd.to_numeric(
    df["streams"].astype(str).str.replace(",", "").str.strip(),
    errors="coerce")

# --- Método 1: IQR ---
Q1 = df["streams"].quantile(0.25)
Q3 = df["streams"].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers_iqr = df[(df["streams"] < lower) | (df["streams"] > upper)]["streams"]
df_no_outliers_iqr = df[(df["streams"] >= lower) & (df["streams"] <= upper)]

total = len(df)
outliers_count = len(outliers_iqr)
remaining = len(df_no_outliers_iqr)
percentage_outliers = (outliers_count / total) * 100

print("=== MÉTODO IQR ===")
print(f"Total de datos originales: {total}")
print(f"Datos eliminados (outliers): {outliers_count}")
print(f"Porcentaje eliminados: {percentage_outliers:.2f}%")

# --- Visualización ---
plt.figure(figsize=(10,4))
sns.boxplot(x=df["streams"])
plt.title("Outliers en 'streams' (Boxplot)")
plt.show()
```

```python
#@title VER ESTRUCTURA
print("Dimensiones del dataset:", df.shape)
print("\nTipos de datos:")
print(df.dtypes.value_counts())

# Revisar las primeras filas
df.head()
```

```python
#@title LIBRERIAS ADICIONALES
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

!pip install xgboost
from xgboost import XGBRegressor

import warnings
warnings.filterwarnings("ignore")
RSEED = 4
```

```python
#@title LIMPIEZA BASICA Y TRANSFORMACION

# Convertir columnas con comas a numéricas (streams y playlist counts)
cols_numeric_commas = ['streams','in_spotify_playlists','in_apple_playlists','in_deezer_playlists']
for c in cols_numeric_commas:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c].astype(str).str.replace(',', '').str.strip(), errors='coerce')

# rellenar 'key' si existe (modo)
if 'key' in df.columns:
    df['key'] = df['key'].fillna(df['key'].mode()[0])

# Drop filas sin target
before = df.shape[0]
df = df.dropna(subset=['streams']).copy()
after = df.shape[0]
print(f"\nFilas con 'streams' disponibles: {after} (eliminadas: {before-after})")

df.head()
```

```python
#@title Selección de variables
# Lista inicial de features candidatas
features = [
    'danceability_%','valence_%','energy_%','acousticness_%',
    'instrumentalness_%','liveness_%','speechiness_%',
    'in_spotify_playlists','in_apple_playlists','in_deezer_playlists',
    'released_year','artist_count']

# Filtrar solo las features que existen en el df
available_features = [f for f in features if f in df.columns]
print("Features candidatas disponibles:", available_features)

# Imputación simple por mediana para features numéricas faltantes
for f in available_features:
    nnull = df[f].isna().sum()
    if nnull > 0:
        df[f] = df[f].fillna(df[f].median())
        print(f"Imputado {nnull} nulos en {f} por mediana")

# Distribución target (por si está muy sesgada)
plt.figure(figsize=(8,4))
sns.histplot(df['streams'], kde=True)
plt.title("Distribución de 'streams' (target)")
plt.show()

# comprobar asimetría; si es muy sesgada luego aplicaremos log1p
print("Skewness de streams:", float(df['streams'].skew()))
```

```python
#@title EDA de correlaciones (justificación de selección)
corr_cols = available_features + ['streams']
plt.figure(figsize=(10,8))
sns.heatmap(df[corr_cols].corr(), annot=True, fmt=".2f", cmap="RdBu_r", center=0)
plt.title("Mapa de correlaciones (features vs streams)")
plt.show()
```

```python
#@title Feature Selection - SelectKBest (f_regression)
# -------------------------
# Escalar antes (SelectKBest usa estadística, scaling ayuda a comparabilidad)
X = df[available_features].copy()
y = df['streams'].copy()

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

k = min(6, X.shape[1])  # quedarnos con 6 o menos si no hay tantas features
selector = SelectKBest(score_func=f_regression, k=k)
selector.fit(X_scaled, y)

selected_idx = selector.get_support(indices=True)
selected_features = [available_features[i] for i in selected_idx]
scores = pd.Series(selector.scores_, index=available_features).sort_values(ascending=False)

print("\nSelectKBest - scores (desc):")
display(scores)
print(f"\n▶ Variables seleccionadas (k={k}): {selected_features}")

# gráfica de los scores
plt.figure(figsize=(8,4))
sns.barplot(x=scores.values, y=scores.index, palette="viridis")
plt.title("Score f_regression por feature")
plt.xlabel("F-score")
plt.show()
```

```python
#@title Preparar datos para modelado

X_model = df[selected_features].copy()
y_model = y.copy()

skewness = float(y_model.skew())
if skewness > 2:
    print(f"\nAdvertencia: target muy sesgado (skew={skewness:.2f}). Se aplicará y = log1p(streams).")
    y_model = np.log1p(y_model)
    transformed_target = True
else:
    transformed_target = False

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_model, y_model, test_size=0.2, random_state=RSEED)
print("\nTamaño train/test:", X_train.shape[0], X_test.shape[0])
```

```python
#@title Modelos a comparar
# -------------------------
models = {"RandomForest": RandomForestRegressor(n_estimators=200, max_depth=10, random_state=RSEED),
    "GradientBoosting": GradientBoostingRegressor(random_state=RSEED),
    "XGBoost": XGBRegressor(random_state=RSEED, objective='reg:squarederror', n_estimators=200)}
```

```python
#@title Entrenar y evaluar (métricas base)
results = {}
cv_scores = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # Si target fue transformado con log1p, invertimos para error en unidades originales (opcional)
    if transformed_target:
        y_test_orig = np.expm1(y_test)
        y_pred_orig = np.expm1(y_pred)
        mae = mean_absolute_error(y_test_orig, y_pred_orig)
        mse = mean_squared_error(y_test_orig, y_pred_orig)
        # r2 en escala transformada y en original puede diferir; reportamos R2 en transformada
        r2 = r2_score(y_test, y_pred)
    else:
        mae = mean_absolute_error(y_test, y_pred)
        mse = mean_squared_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)

    results[name] = {"MAE": mae, "MSE": mse, "R2": r2}

    # Cross-val R2 (usar 5 folds sobre todo X_model)
    try:
        cv = cross_val_score(model, X_model, y_model, cv=5, scoring='r2', n_jobs=-1)
        cv_scores[name] = {"cv_mean": cv.mean(), "cv_std": cv.std()}
    except Exception as e:
        cv_scores[name] = {"cv_mean": np.nan, "cv_std": np.nan}

  # Mostrar resultados
results_df = pd.DataFrame(results).T
cv_df = pd.DataFrame(cv_scores).T
display(results_df.round(4))
display(cv_df.round(4))




```

```python
#Visualizaciones comparativas
# -------------------------
fig, axes = plt.subplots(1, 3, figsize=(18, 5))  # 1 fila, 3 columnas

#Comparación de R² por modelo
sns.barplot(ax=axes[0], x=results_df.index, y=results_df["R2"], palette="viridis")
axes[0].set_title("Comparación R² por modelo")
axes[0].set_ylim(min(results_df["R2"].min() - 0.05, 0), results_df["R2"].max() + 0.05)
axes[0].set_ylabel("R²")
axes[0].set_xlabel("Modelos")
axes[0].tick_params(axis='x', rotation=45)

#Elegir mejor modelo
best_model_name = results_df["R2"].idxmax()
best_model = models[best_model_name]
print(f"\n▶ Mejor modelo (según R2 en test): {best_model_name}")

# Predicciones del mejor modelo
y_pred_best = best_model.predict(X_test)
if transformed_target:
    y_test_plot = np.expm1(y_test)
    y_pred_plot = np.expm1(y_pred_best)
else:
    y_test_plot = y_test
    y_pred_plot = y_pred_best

#Gráfico de Predicción vs Real
sns.scatterplot(ax=axes[1], x=y_test_plot, y=y_pred_plot, alpha=0.6, color="teal")
axes[1].plot(
    [y_test_plot.min(), y_test_plot.max()],
    [y_test_plot.min(), y_test_plot.max()],
    'k--', linewidth=1)
axes[1].set_xlabel("Valor real (streams)")
axes[1].set_ylabel("Predicción (streams)")
axes[1].set_title(f"Predicción vs Real - {best_model_name}")

#Distribución de Residuales ---
resid = y_test_plot - y_pred_plot
sns.histplot(ax=axes[2], data=resid, kde=True, color='purple')
axes[2].set_title(f"Distribución de residuales - {best_model_name}")
axes[2].set_xlabel("Residual (real - predicho)")
axes[2].set_ylabel("Frecuencia")

plt.tight_layout(w_pad=8.0)
plt.show()
```

```python
 #@title Importancia de features
# -------------------------
if hasattr(best_model, "feature_importances_"):
    fi = pd.Series(best_model.feature_importances_, index=selected_features).sort_values(ascending=True)
    plt.figure(figsize=(8,4))
    fi.plot(kind="barh", color="indigo")
    plt.title(f"Feature Importances - {best_model_name}")
    plt.xlabel("Importancia relativa")
    plt.show()
    print("\nImportancias (desc):")
    display(fi.sort_values(ascending=False))
else:
    print("\nEl modelo no expone 'feature_importances_' (por ejemplo, algunos wrappers).")
```

```python
#@title Validación cruzada final del mejor modelo (R2)
# -------------------------
try:
    final_cv = cross_val_score(best_model, X_model, y_model, cv=5, scoring='r2', n_jobs=-1)
    print(f"\nCross-val R2 (5-fold) del mejor modelo {best_model_name}: mean={final_cv.mean():.4f}, std={final_cv.std():.4f}")
except Exception as e:
    print("No se pudo ejecutar cross_val_score en el mejor modelo:", e)
```
